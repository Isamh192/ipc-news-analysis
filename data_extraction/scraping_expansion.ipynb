```python
import pandas as pd
import numpy as np
import feedparser
import re
from datetime import datetime
```

## Obtención de Noticias

```python
url_expansion = 'https://expansion.mx/rss/economia'
```

```python
# Definición de patrones regulares
pat = re.compile(r'rss/([a-z]+)(?=\.xml|$)')
pat2 = re.compile(r'//([a-z]+)\.mx|www\.([a-z]+)\.com')
```

```python
def corpus_create(URL: str) -> pd.DataFrame:
    feed = feedparser.parse(URL)

    titles_news = []
    sumary_news = []
    url_news = []
    date_news = []
    date_news_form = []
    section_news = []
    source_news = []

    seccion = pat.findall(URL)
    source = pat2.findall(URL)

    seccion = seccion[0] if seccion else None
    source = source[0][0] if source[0][0] else source[0][1]

    for i in range(len(feed.entries)):
        titulo = feed.entries[i].title
        contenido = feed.entries[i].summary
        fecha = feed.entries[i].published
        links = feed.entries[i].link

        url_news.append(links)
        titles_news.append(titulo)
        sumary_news.append(contenido)
        date_news.append(fecha)
        section_news.append(seccion)
        source_news.append(source)

    for j in range(len(date_news)):
        object_date = datetime.strptime(date_news[j], '%a, %d %b %Y %H:%M:%S %Z')
        date_form = object_date.strftime('%d/%m/%Y')
        date_news_form.append(date_form)

    corpus = pd.DataFrame({
        'fuente': source_news,
        'titulo': titles_news,
        'contenido': sumary_news,
        'seccion': section_news,
        'liga': url_news,
        'fecha': date_news_form,
    })

    return corpus
```

```python
corpus = corpus_create("https://expansion.mx/rss/economia")
```

## Filtrado de Noticias 

Para filtrar las noticias de manera efectiva, es necesario aplicar técnicas de extracción de palabras clave que permitan identificar los temas principales de cada una.
Teniendo esto en cuenta, se aplicaron dos algoritmos de extracción de palabras.

### TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import spacy
```

```python
nlp = spacy.load('es_core_news_sm') 
```

```python
def preprocess_text_simple(text):# Normalización
    text = text.lower()
    text = ''.join([char for char in text if char.isalpha() or char.isspace()])
    doc = nlp(text)
    lemmatized = [token.lemma_ for token in doc if not token.is_punct and not token.is_space]
    return ' '.join(lemmatized)
```

```python
df['titulo_preprocesado'] = df['titulo'].apply(preprocess_text_simple)
df['contenido_preprocesado'] = df['contenido'].apply(preprocess_text_simple)

df['texto_completo'] = df['titulo_preprocesado'] + ' ' + df['contenido_preprocesado']
```

```python
spanish_stopwords = [
    'de', 'la', 'que', 'el', 'en', 'y', 'a', 'los', 'del', 'las', 'se', 'por', 'con', 'un', 'para', 'es', 
    'como', 'más', 'o', 'su', 'una', 'sus', 'al', 'lo', 'ya', 'este', 'todo', 'pero', 'si', 'también', 
    'fue', 'está', 'han', 'ser', 'puede', 'ha', 'te', 'esta', 'cuando', 'muy', 'nos', 'sobre', 'uno', 
    'entre', 'hasta', 'desde', 'ni', 'contra', 'otros', 'ese', 'esto', 'tiempo'
]
```

```python
def filter_nouns_spacy(text): # Filtrar sustantivos
    doc = nlp(text)
    return " ".join([token.text for token in doc if token.pos_ == 'NOUN' and token.text not in spanish_stopwords])


df['texto_filtrado'] = df['texto_completo'].apply(filter_nouns_spacy)
```

```python
# Aplicar TF-IDF
vectorizer = TfidfVectorizer()
X_tfidf_sustantivos = vectorizer.fit_transform(df['texto_filtrado'])
X_tfidf_dense_sustantivos = X_tfidf_sustantivos.toarray()
feature_names = vectorizer.get_feature_names_out()

# Extraer palabras clave
def get_keywords_for_news(news_vector, top_n=10):
    sorted_indices = news_vector.argsort()[::-1]
    top_words_idx = sorted_indices[:top_n]
    top_words = [feature_names[i] for i in top_words_idx]
    return " ".join(top_words)
```

```python
df_tfidf = pd.DataFrame()
df_tfidf['titulo'] = df['titulo']
df_tfidf['palabras_clave'] = [
    get_keywords_for_news(X_tfidf_dense_sustantivos[i], top_n=10)
    for i in range(X_tfidf_dense_sustantivos.shape[0])
]
df_tfidf
```

### **Ocupando TextRank**

```python
import networkx as nx
from itertools import combinations
from collections import Counter
```

```python
def preprocess_text(text):
    text = text.lower()
    doc = nlp(text)
    words = [token.lemma_ for token in doc if token.pos_ in ['NOUN', 'PROPN', 'ADJ'] and not token.is_stop and token.is_alpha]
    return words
```

```python
def preprocess_text_lemma(text):
    doc = nlp(text.lower())
    words = []
    for token in doc:
        if token.is_alpha and not token.is_stop and token.pos_ in ['NOUN', 'PROPN', 'ADJ']:
            words.append((token.lemma_, token.pos_)) 
    return words
```

```python
def textrank_keywords(words, top_n=10, window_size=5):
    # Crear lista de pares de coocurrencia
    graph = nx.Graph()
    for i in range(len(words)):
        for j in range(i+1, i+window_size):
            if j >= len(words):
                break
            graph.add_edge(words[i], words[j])
    
    # Algoritmo PageRank
    ranks = nx.pagerank(graph)
    
    sorted_words = sorted(ranks.items(), key=lambda x: x[1], reverse=True)
    top_keywords = [word for word, score in sorted_words[:top_n]]
    return " ".join(top_keywords)


```

```python
df['texto_completo'] = df['titulo'].fillna('') + ' ' + df['contenido'].fillna('')
```

```python
# Crear nuevo DataFrame con resultados
df_texTrank = pd.DataFrame()
df_texTrank['titulo'] = df['titulo']
df_texTrank['palabras_clave'] = df['texto_completo'].apply(lambda x: textrank_keywords(preprocess_text(x), top_n=6))
df_texTrank 
```

```python
indicadores_macro = [
    "inflación", "oferta", "demanda", "recesión", "precios", "política monetaria", "tasa de interés",
    "tipo de cambio", "salarios", "poder adquisitivo", "consumo", "producción", "costo de vida",
    "crecimiento económico", "expectativas", "impuestos", "gastos gubernamentales", "aranceles",
    "cetes", "petroleo", "indice dolar", "tiee", "tarifas", "materias primas", "estacionalidad",
    "ajustes salariales", "inversión", "política fiscal", "desempleo", "confianza del consumidor",
    "financiamiento", "desaceleración"
]
```

```python
empresas_ipc = [
    "alfa", "alsea", "américa móvil", "arca continental", "banorte", "bajío", "becle", "bolsa mexicana",
    "cemex", "chedraui", "coca-cola femsa", "vesta", "liverpool", "femsa", "genomma lab", "gentera", 
    "gruma", "oma", "gap", "asur", "bimbo", "carso", "gcc", "elektra", "gmexico", "inbursa", 
    "kimberly", "megacable", "orbia", "peñoles", "infraestructura", "quálitas", "regional", "walmart",
    "banxico", "banco de méxico"
]
```

```python
lista_claves = indicadores_macro + empresas_ipc
lista_claves = [x.lower() for x in lista_claves]

df_combinado = pd.merge(df_texTrank , df, on="titulo", how="inner")

def extraer_coincidencias(titulo, palabras_clave):
    texto = f"{titulo} {palabras_clave}".lower()
    return ', '.join(sorted({palabra for palabra in lista_claves if palabra in texto}))

df_combinado['coincidencias'] = df_combinado.apply(
    lambda row: extraer_coincidencias(row['titulo'], row['palabras_clave']), axis=1
)

df_filtrado = df_combinado[df_combinado['coincidencias'].str.strip() != ""]

df_filtrado[['titulo', 'palabras_clave', 'coincidencias', 'contenido', 'liga', 'fecha']]
```
